# STUDENT RESULTS DATABASE MANAGEMENT

```python
import numpy as np
```

## Creating a dataset

```python
dataset = np.random.randint(0,101, size=(100,5))
print("Students marks dataset:",dataset)
```

## Adding Labels

```python
student_ids = np.arange(1, 101)
subject_names = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
```

## Sum of total marks

```python
total_marks = np.sum(dataset, axis=1)
print("Sum of each student's marks:", total_marks)
```

## Top 5 Students

```python
top_5_indices = np.argsort(total_marks)[::-1][:5]
top_5_scores = total_marks[top_5_indices]
print("Top 5 Students:",top_5_scores)
```

## Subject Wise Statistics


```python
avg_marks_per_subject = np.mean(dataset,axis=0)
print("Average marks per subject:",avg_marks_per_subject)
```

```python
min_marks_per_subject = np.min(dataset, axis=0)
print("Minimum marks per subject:", min_marks_per_subject)
```

```python
max_marks_per_subject = np.max(dataset, axis=0)
print("Maximum marks per subject:", max_marks_per_subject)
```

```python
topper_per_subject = np.argmax(dataset, axis=0)
print("Topper per subject:", topper_per_subject)
```

## Output Insights

```python
print("Top 5 Students (by total marks):")
for i in range(5):
    print(f"Student {top_5_indices[i]+1}: {top_5_scores[i]} marks")

print("Subject-wise Averages:", avg_marks_per_subject)
print("Subject-wise Maximums:", max_marks_per_subject)
print("Subject-wise Minimums:", min_marks_per_subject)

print("\nTopper in Each Subject:")
for i, idx in enumerate(topper_per_subject):
    print(f"{subject_names[i]}: Student {idx+1} with {dataset[idx][i]} marks")
```
